```python
import pandas as pd
import numpy as np
```

```python
alternatives = pd.read_csv('dataset.csv')
alternatives
```

## Input the Criteria

```python
harga = int(input('Harga? '))
jarak_tempuh = int(input('Jarak tempuh? '))
jumlah_wahana = int(input('Jumlah wahana? '))
rating = int(input('Rating? '))

dict_criteria = {'Harga': harga, 'Jarak Tempuh (Km)': jarak_tempuh, 'Jumlah Wahana': jumlah_wahana, 'Rating': rating}
```

## Weight Priority from Criteria

```python
total = 0
for key, value in dict_criteria.items():
    total += value

criteria = [ (key, (value/total)) for key, value in dict_criteria.items() ]
criteria = dict(criteria)

print(criteria)
```

```python
criteria['Harga']
```

## Calculate Si Vector for each Criteria and Alternatives

```python
si_vector = alternatives.copy()
```

```python
si_vector['Harga (Vektor Si)'] = si_vector['Harga'].apply(lambda x: x**(-criteria['Harga']))
si_vector['Jarak Tempuh (Vektor Si)'] = si_vector['Jarak Tempuh (Km)'].apply(lambda x: x**(-criteria['Jarak Tempuh (Km)']))
si_vector['Jumlah Wahana (Vektor Si)'] = si_vector['Jumlah Wahana'].apply(lambda x: x**(criteria['Jumlah Wahana']))
si_vector['Rating (Vektor Si)'] = si_vector['Rating'].apply(lambda x: x**(criteria['Rating']))
```

```python
si_vector
```

```python
si_vector['Total Si'] = si_vector.iloc[:, 5:].apply(lambda row: row['Harga (Vektor Si)'] * \
                   row['Jarak Tempuh (Vektor Si)'] * row['Jumlah Wahana (Vektor Si)'] * row['Rating (Vektor Si)'], axis=1)
```

```python
si_vector
```

```python
total_s = np.sum(si_vector['Total Si'])
total_s
```

## Calculate Vi Vector

```python
v_vector = si_vector.copy()
```

```python
v_vector['Vektor Vi'] = v_vector['Total Si'].apply(lambda row: (row/total_s))
```

```python
v_vector
```

## Sort Values based on Vi Vector

```python
sorted_vi = v_vector.sort_values(by=['Vektor Vi'], ascending=False)
```
